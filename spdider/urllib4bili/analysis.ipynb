```
# [' coin', ' danmaku', ' favorite', ' like', ' reply', ' share',' view']
# 假定 score = p0  + p1*coin + p2*danmaku + p3*favorite + p4*like + p5*reply + p6*share + p7*view + err
```
算法：

$$
X = x_i^J |size->(m*8) \\
\Theta = [\theta_1, \theta_2, ... \theta_8] \\
h(\Theta)= X*\Theta^T |size->(8*1)\\
J(\Theta) = \frac{1}{2m}\sum_{i=1}^{m}(h^i(\Theta)-y^i)^2 \\
repeat: \theta_j := \theta_j - \alpha * \frac{{\partial J(\theta)}}{\partial \theta_j} \\
\frac{{\partial J(\Theta)}}{\partial \theta_j} = \frac{1}{m}\sum_{i=1}^{m}((h(\Theta)-y^i)*x^i_j) (j=0,1,2 ... 8)
$$

计算：  
1. 归一化
2. 给定参数初始值，运行算法
3. 输出参数

````python
tmp=dat[['aid',' coin']].copy()
tmp.loc[:,' coin'] = 1
tmp.columns=['aid','base']
theta = [1,1,1,1,1,1,1]
tmp = pd.merge(tmp,dat,on='aid')

keys=['base',' coin', ' danmaku', ' favorite', ' like', ' reply', ' share',' view', ' score']
tmp = tmp[keys]
new_keys = ['base','coin', 'danmaku', 'favorite', 'like', 'reply', 'share','view', 'score'] # 去掉空格
tmp.columns=new_keys
tmp.head()
````

````python
tmp.describe()
````

````python
## 归一化
# (x-mean)/(max-min)
tmp = (tmp-tmp.mean())/(tmp.max()-tmp.min())
tmp['base'] = 1
tmp.head()
````

````python
feature_keys = ['base','coin', 'danmaku', 'favorite', 'like', 'reply', 'share','view']
feature_value = tmp[feature_keys].copy()
score = tmp['score'].copy()
theta = [1.,1.,1.,1.,1.,1.,1.,1.]
````

````python
hpy=feature_value.dot(theta)
````

````python
diff = hpy-score
sum(feature_value['danmaku']*diff)
````

````python
i = 0
alpha = 0.01
for key in feature_keys:
    print(key)
    theta[i] = theta[i] - alpha*sum(feature_value[key]*diff)/len(diff)
    i += 1
````

````python
def bgd(var_dat,res_dat,alpha,theta=None):
    """
    var_dat: 变量数据集
    res_dat: 因变量数据集
    alpha: learning rate 学习率参数
    theta: 变量系数的初始值, 默认为1
    """
    # batch_gradient _descent 批量梯度下降
    if theta == None:
        theta = [i/i for i in range(1,var_dat.shape[1]+1)]
    diff=var_dat.dot(theta)-res_dat
    i=0
#     求各项的参数
    for key in var_dat.columns:
#         print(key)
        theta[i] = theta[i]-alpha*sum(var_dat[key]*diff)/len(diff)
        i+=1
    cost = sum(diff**2)/len(diff)/2 # 损失函数值
    return (theta,cost)  # 返回新的参数列表及此时的损失函数值
# new_theta,cost = bgd(feature_value,score,alpha)
````
